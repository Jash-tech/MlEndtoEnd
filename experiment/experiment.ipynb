```python
num_pipeline=Pipeline (
    steps=[
        ("imputer",SimpleImputer()),
        ("scaler",StandardScaler())
    ]
)
```

Lasso Regression:

Lasso: This is a type of linear regression that includes a regularization term. The regularization term helps to prevent overfitting by penalizing large coefficients, effectively setting some of them to zero. This is useful for feature selection because it automatically excludes less important features.

Parameters:

alpha=5e-5: This sets the regularization strength. A smaller alpha means less regularization, and a larger alpha means more regularization. Here, 5e-5 is a very small value, indicating light regularization. This means the model will only slightly penalize large coefficients.

random_state=1: This ensures reproducibility. By setting a seed (in this case, 1), you ensure that the results are the same every time you run the code. This is useful for debugging and comparing models.

max_iter=1e+6: This sets the maximum number of iterations the algorithm will run to find the optimal solution. 1e+6 is a large number (1,000,000), ensuring that the algorithm has plenty of opportunities to converge to a solution.

Fitting the Model:

.fit(X_train_scale, y_train): This trains the Lasso regression model using the provided training data.
X_train_scale: These are the input features for training, usually scaled to ensure all features contribute equally to the model.
y_train: These are the target values corresponding to the input features. The model learns the relationship between X_train_scale and y_train during training.

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
import pandas as pd

# Establish the Lasso (L1) Regularisation model that will perform feature selection.
lasso = Lasso(alpha=1e-4, random_state=1, max_iter=1000).fit(X_train, y_train)
model = SelectFromModel(lasso, prefit=True)

# Transform the training data to keep only selected features
X_train_l1 = model.transform(X_train)

# Transform the test data in the same way
X_test_l1 = model.transform(X_test)

# Get the mask of selected features
selected_mask = model.get_support()

# Get the selected feature names
selected_feature_names = X_train.columns[selected_mask]
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor

import numpy as np
```

```python
model={'Random Forest':RandomForestRegressor(),'LR':LinearRegression(),'DT':DecisionTreeRegressor(min_samples_leaf= 1, min_samples_split=5)}
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    r2 = r2_score(actual, pred)
    return rmse, mae, r2
```

```python
for m in range(len(list(model))):
    mod=list(model.values())[m]
    mod.fit(X_train,y_train)
    prediction=mod.predict(X_test)
    RMSE,MAE,R2=eval_metrics(y_test,prediction)
    print("Model performance:",mod)
    print("RMSE",RMSE)
    print("MAE",MAE)
    print("R2",R2)
   
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

# Define the model
dr = DecisionTreeRegressor(random_state=1)

# Define the parameter grid
param_grid = {
    
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Set up the GridSearchCV
grid_search = GridSearchCV(estimator=dr, param_grid=param_grid, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')

# Fit the model
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_

print("Best parameters found: ", best_params)
```
